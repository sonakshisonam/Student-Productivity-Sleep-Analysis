```python
Model Training
```

```python
X = df[['Study_Hours', 'Screen_Time', 'Sleep_Hours', 'Caffeine', 'Physical_Activity']]
y_sleep = df['Sleep_Quality']
y_productivity = df['Productivity_Score']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Sleep_Model = LinearRegression()
Sleep_Model.fit(X, y_sleep)
```

```python
Productivity_Model = LinearRegression()
Productivity_Model.fit(X, y_productivity)
```

```python
print("Sleep Model Coefficients:")
for f, c in zip(X.columns, Sleep_Model.coef_):
    print(f, ":", c)

print("\nProductivity Model Coefficients:")
for f, c in zip(X.columns, Productivity_Model.coef_):
    print(f, ":", c)
```

```python
Saving the trained model
```
